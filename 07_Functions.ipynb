## Syntax for defining a function:

## A function that does not accept arguments

#### This is an example of a print message.   


```python
def print_welcome():
    print("Welcome to the cool app")
    print()

# How to call the function
print_welcome()
```

<hr>

## A function that has one argument

```python
def print_welcome(message):
    print(message)
    print()


message = "Welcome to the cool app"

# How to call the function
print_welcome(message)
```

## A function that has two arguments

```python
def calculate_miles_per_gallon(miles_driven, gallons_used):
    mpg = miles_driven / gallons_used
    mpg = round(mpg, 2)
    return mpg


miles = 500
gallons = 14
mpg = calculate_miles_per_gallon(miles, gallons)

# How to call the function
print_welcome(mpg)
```
